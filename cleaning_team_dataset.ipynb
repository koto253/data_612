```python
# dropping divID it cointains 52% missing values
# dropping DivWin it cointains 53% missing values
# dropping WCWin: it cointains 75% missing values
teams = teams.drop(['divID', 'DivWin', 'WCWin'], axis=1)
```

```python
teams.count()
```

```python
columns_to_refill = ['yearID', 'Rank', 'Ghome', 'BPF', 'PPF']
column_means = teams[columns_to_refill].mean()
teams[columns_to_refill] = teams[columns_to_refill].fillna(column_means)
teams.count()
```

```python
teams.rename(columns={'yearID': 'year',
                     'lgID': 'state',
                      'name': 'team'}, 
            inplace=True)
```

```python
# dropping the rest of missing vales
teams.dropna()
```

```python
# set index by team columns
# teams.set_index('team', inplace=True)
teams.reset_index(inplace=True)
teams
```

```python
teams.shape[0] - teams.count() 
```

```python
# dropping duplicates
teams.drop_duplicates()
teams # it seems like this dataset does not contain any duplicates
```
