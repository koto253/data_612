```python
# loading pandas library
import pandas as pd
```

```python
# loading the dataset of state drug utilization 
state_drug = pd.read_csv('datasets/state_drug_utilization.csv')

# displaying the first 5 rows
state_drug.head()
```

```python
# making a subset dataset from the original dataset
df = state_drug.iloc[10:21 , [6, 2, 8, 11, 12, 5]]
df.head()
```

```python
df.info()
```

```python
# converting 'Package Size column from int64 to categorical type
df['Package Size'] = df['Package Size'].astype('category')
df.info()
```

```python
# converting year datatype from object to datetime
df['Year'] = pd.to_datetime(df['Year'])
```

```python
df.dtypes
```

```python
# converting 'number of prescription' column from int64 to string type
df['Number of Prescriptions'] = df['Number of Prescriptions'].astype(str)
```
