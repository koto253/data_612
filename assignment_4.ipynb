```python
# import Libraries
import pandas as pd
import seaborn as sns
```

```python
stocks_yahoo = pd.read_csv("stocks_yahoo.csv")
# stocks_yahoo
```

```python
# using scatter plot
# to observe the price_at_close and price_after_hours

sns.relplot(data=stocks_yahoo, x='price_at_close', y='price_after_hours')

# Use a SCATTER PLOT to examine the relationship between 2 continous variables
# can use it, to inspect for cause-and-effect analysis or 
# To detec correlations
# like the graph below. 
```

```python
# line graph 
sample_stock = stocks_yahoo.sample(10)
sns.relplot(data=sample_stock, x='earnings date', y='price_at_close', kind='line')

# Line graph is ideal for showing trends over time or continuous data
```

```python
# bar gragh using the sample_stock dataset
sns.barplot(sample_stock, x='company_name', y='price_at_close', hue='company_name')

# We use bar graph when we want to compare DISCRETE catergories
```
