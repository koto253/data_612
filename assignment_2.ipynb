```python
# sample 20 rows from the data  
# Since data is large, I decided to sample a small portion of the data
sample_df = df.sample(20)
```

```python
# group the data by Product Name and calculate the mean number of prescriptions 
# save the result in a new dataframe called grouped 
# sort the values in descending order  
grouped = round(sample_df.groupby(['Product Name'])['Number of Prescriptions'].mean())
grouped.sort_values(ascending=False)
```
